$similarityscore_{document} = \frac{\sum_n^4 n F1(n)}{\sum_n^4 n}$

```python
import json
import spacy
import textacy
nlp = spacy.load("en_core_web_sm")
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
def get_doc_ngrams(doc,min_freq=1):
    unigrams = [str(i) for i in set(textacy.extract.basics.ngrams(doc, 1, min_freq=min_freq))]
    bigrams = [str(i) for i in set(textacy.extract.basics.ngrams(doc, 2, min_freq=min_freq))]
    trigrams = [str(i) for i in set(textacy.extract.basics.ngrams(doc, 3, min_freq=min_freq))]
    quadgrams = [str(i) for i in set(textacy.extract.basics.ngrams(doc, 4, min_freq=min_freq))]
    return [unigrams,bigrams,trigrams,quadgrams]
```

```python
source_document_raw = ''
with open('queries.txt','r') as f:
    for l in f:
        source_document_raw += l

        
source_document = nlp(source_document_raw.lower())
source_ngrams = get_doc_ngrams(source_document)
```

```python
j = 0
# 1 extract ngrams
# 2 extract TP FP and FN
# TP is the len of set intersection of target document and source document
# FP = len(target document) - len(intersection)
# FN = len(source document) - len(intersection)
with open('en.txt','r') as f:
    for l in f:
        target_document_raw = l.strip().lower()
        target_document = nlp(target_document_raw.lower())
        target_ngrams = get_doc_ngrams(target_document)
        score = 0
        for i in range(4):
            TP = len(intersection(source_ngrams[i],target_ngrams[i]))
            FP = len(target_ngrams[i]) - TP
            FN = len(source_ngrams[i]) - TP
            F1 = 0
            if TP > 0:
                P = TP/(TP+FP)
                R = TP/(TP+FN)
                F1 = (2*P*R)/(P+R)
            score += F1 * (i + 1)
        score /= 10 # Denominator is the sum of frequencies so 1+2+3+4=10
        scores[j] = score
        j += 1
        
top_percent = int(len(scores)/1000)
top_items = sorted(scores,key=scores.get,reverse=True)[:top_percent]

j=0
with open('en.txt','r') as f:
    with open('stories-cc.txt','w') as w:
        for l in f:
            if j in top_items:
                w.write(l)
            j+= 1

```
